```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from matplotlib.backends.backend_pdf import PdfPages
%matplotlib inline
import json
```

```python
with open('word_list.txt') as data_file:    
    data = json.load(data_file)
```

```python
words = data.keys()
```

```python
with PdfPages("testpdf.pdf") as pdf:
    fig = plt.figure(figsize=(6, 10))
    fig.subplots_adjust(wspace=0.05, hspace=0.05, bottom=0.005, top=0.995, left=0.01, right=0.99)

    for ii in range(len(words)):
        ax = plt.subplot(3,2,(ii+1))
        ax.fill_between([0, 1], [0.75, 0.75], [1.0, 1.0], color="blue")
        ax.plot([0, 1], [0.75, 0.75], linewidth=2, color="k")
        font = FontProperties().copy()
        font.set_weight("bold")
        font.set_size(12)
        ax.text(0.5, 0.85, words[ii], transform = ax.transAxes,
                horizontalalignment="center", color="white", fontproperties=font)
        ax.text(0.5, 0.05, data[words[ii]][0], transform = ax.transAxes,
                horizontalalignment="center", color='k', fontsize=12)
        ax.text(0.5, 0.2, data[words[ii]][1], transform = ax.transAxes,
                horizontalalignment="center", color='k', fontsize=12)
        ax.text(0.5, 0.35, data[words[ii]][2], transform = ax.transAxes,
                horizontalalignment="center", color='k', fontsize=12)
        ax.text(0.5, 0.5, data[words[ii]][3], transform = ax.transAxes,
                horizontalalignment="center", color='k', fontsize=12)
        ax.text(0.5, 0.65, data[words[ii]][4], transform = ax.transAxes,
                horizontalalignment="center", color='k', fontsize=12)
        ax.set_xlim(0, 1)
        ax.set_ylim(0, 1)
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        ax.set_xticks([])
        ax.set_yticks([])
    pdf.savefig()

    fig = plt.figure(figsize=(6, 10))
    fig.subplots_adjust(wspace=0.05, hspace=0.05, bottom=0.005, top=0.995, left=0.01, right=0.99)

    for ii in range(len(words)):
        ax = plt.subplot(3,2,(ii+1))
        ax.fill_between([0, 1], [0.75, 0.75], [1.0, 1.0], color="blue")
        ax.plot([0, 1], [0.75, 0.75], linewidth=2, color="k")
        font = FontProperties().copy()
        font.set_weight("bold")
        font.set_size(12)
        ax.text(0.5, 0.85, words[ii], transform = ax.transAxes,
                horizontalalignment="center", color="white", fontproperties=font)
        ax.text(0.5, 0.05, data[words[ii]][0], transform = ax.transAxes,
                horizontalalignment="center", color='k', fontsize=12)
        ax.text(0.5, 0.2, data[words[ii]][1], transform = ax.transAxes,
                horizontalalignment="center", color='k', fontsize=12)
        ax.text(0.5, 0.35, data[words[ii]][2], transform = ax.transAxes,
                horizontalalignment="center", color='k', fontsize=12)
        ax.text(0.5, 0.5, data[words[ii]][3], transform = ax.transAxes,
                horizontalalignment="center", color='k', fontsize=12)
        ax.text(0.5, 0.65, data[words[ii]][4], transform = ax.transAxes,
                horizontalalignment="center", color='k', fontsize=12)
        ax.set_xlim(0, 1)
        ax.set_ylim(0, 1)
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        ax.set_xticks([])
        ax.set_yticks([])
    pdf.savefig()

        
# plt.show()
```
